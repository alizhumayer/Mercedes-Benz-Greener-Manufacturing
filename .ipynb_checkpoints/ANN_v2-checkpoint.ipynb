```python
import tensorflow as tf

from keras.layers import Dense, Activation

from keras.models import Sequential

import numpy as np

```

```python
np.random.seed(7)
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io

df = pd.read_csv(io.StringIO(uploaded['letter_data.csv'].decode('utf-8')))
df
```

```python

x = np.array(df[:19999].drop(['T','2','8','3','5','1','0.1','0.2'], 1))
from sklearn.preprocessing import StandardScaler
sc=StandardScaler(with_mean=True,with_std=True)

x = sc.fit_transform(x)
train_x = x[:18000]
test_x = x[18000:]
```

```python
# x
onehot=pd.get_dummies(df, columns='T')
# onehot.to_csv('onehot')
onehot
a = onehot.columns[onehot.columns.str.startswith('T')]

y = onehot.loc[:,a]
train_y = y[:18000]
test_y = y[18000:]
```

```python
 from keras import regularizers
  
 model=Sequential()
```

```python
model.add(Dense(128,input_dim=9, activation='sigmoid', kernel_regularizer=regularizers.l2(0.0001))) #dense class is used for fully connected networks; input_dim defines # of input variables it has 12 neurons
```

```python
model.add(Dense(64, activation='sigmoid', kernel_regularizer=regularizers.l2(0.0001))) #relu is activation function for the first 2 layers also it has 8 neurons
```

```python
model.add(Dense(32, activation='sigmoid'))
```

```python
model.add(Dense(16, activation='sigmoid'))
```

```python
model.add(Dense(26, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(train_x, train_y, nb_epoch=1000, batch_size=50, verbose=2)
```

```python
from sklearn import metrics

expected = test_y
predicted = model.predict(test_x)
b = np.zeros_like(predicted)
b[np.arange(len(predicted)), predicted.argmax(1)] = 1
```

```python

# # # # summarize the fit of the model
print(metrics.classification_report(expected, b))
```

```python
print(metrics.confusion_matrix(expected.values.argmax(axis=1), b.argmax(axis=1)))
```

```python
print(metrics.accuracy_score(expected,b))
```
